```r
library(emmeans)
library(mcprofile)
library(base)
library(tidyverse)
```

```r
WHO_COVID <- read.csv("WHO-COVID-19-global-data.csv")
```

```r
WHO_Lastest_COVID <- read_csv("WHO-COVID-19-global-table-data.csv")
```

```r
WHO_Vaccination_COVID <- read_csv("vaccination-data.csv")
```

```r
head(WHO_Lastest_COVID)
head(WHO_Vaccination_COVID)
```

```r
covid <- WHO_Lastest_COVID %>% select(c(1,3,4,8,9))%>% 
                      arrange(Name)
```

```r
vax <- WHO_Vaccination_COVID %>% select(c(1,6,7,8,9,10,11,14,15,16)) %>%
                            arrange(COUNTRY)
```

```r
covid$Name
```

```r
cleaned <- covid %>% inner_join(vax, by = c("Name" = "COUNTRY"))
```

```r
write_csv(cleaned, path = 'covid_case_and_vax.csv')
```
